# Лекция 5 - Обнаружение и обработка выбросов

## Определение

*Выброс* - точка данных, которая значительно отклоняется от исходных

Виды:

* Глобальные (истинные)
* Контекстные

## Методы обнаружений выбросов

* Статистические
    + Z-оценка
    + Межквартильный размах (IQR)
* Методы машинного обучения
    + Основанные на расстоянии
        - Метод ближайших соседей (KNN)
        - Коэф. локального выброса (LOF)
    + Основанные на кластеризации
        - Пространственная (DBSCAN)
        - Иерархическая
    + Изолирующий лес
    + Одноклассовые машины опорных векторов

* Трансформация
    + Масштабирование
    + Винсоризация - замена выброса ближейшими к нему данными
    + Логарифмические преобразования
* Надёжная оценка - использовать алгоритм, менее чувстсвительный к выбросам
* Моделирование выбросов

## Z-оценка

Z-оценка показывает, насколько точка отличается от среднего значения

Z = (X - a)/s, где a - среднее значение, s - стандартное отклонение
s = sqrt(sum(x_i - X, i=1..N)^2/N)

## Межквартильный размах

* Нижнее = среднее - 3 станд. отклон.
* Верхнее = среднее + 3 станд. отклон.
* IQR = 75% - 25%
* Нижнее = первый квартал - 1.5*IQR
* Верхнее = третий квартал + 1.5*IQR

## Преобразования

* Стандартизация = (x_j - M_i)/s_j
* Диапазонное шкалирование = (x_i - min(x_j))/(max(x_j)-min(x_j))
* Нормализация средним = (x_j - a)/(max(x_j)-min(x_j))
