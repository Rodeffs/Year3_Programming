# ЛЕКЦИЯ 3 - Аугментация данных (data augmentation)

**Аугментация данных** - это методика создания дополнительных данных из имеющихся

## Типы преобразования изображений:

1. Геометрические преобразования (увеличение, обрезка, поворот)
2. Искажение цвета
3. Фильтры
4. Смешивание изображений (т.е. изменение интенсивности RGB канала)
5. Удаление информации
6. Комбинации всего вышеперечисленного

## Применяются следующие библиотеки для работы с нейросетями (и для аугментации изображений):

* Google - Tensor Flow
* Meta - PyTorch

* NumPy (активно используется в sci-kit image)

## Модули scikit-image:

- color
- data
- draw
- exposure - интенсивность изображения
- feature - вытаскивает из изображения определённые части/признаки
- filters
- graph
- io - ввод/вывод, чтение/сохранение изображений
- measure - изменение свойств изображений
- metrics
- morphology
- restoration
- segmentation
- transform
- util

````python
# Работа с scikit-image

from scimage import data, io

import matplotlib.pyplot as plt


image = data.coffee()  # чашка кофе

# Отображение изображения, 2 способа:

io.imshow(image)  # для интерактивной среды
io.show(image)  # для неинтерактивной среды


# Отображение через matplotlib

img = data.coins()  # монетки

print("size:", img.shape)  # вывод размера

plt.imshow(img, cmap="gray")  # вывод изображения через matplotlib с некоторыми параметрами


# Чтение 1 изображения

img = io.imread("path/to/image")

# Чтение нескольких изображений сразу

ims = io.imread_collection("*.jpg")  # это будет массив


# Отображение нескольких изображений сразу

io.imgshow_collection(ims)


# Сохранение изображений

io.imgsave(image)
````
